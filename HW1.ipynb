```python
import nltk, re
import glob
import os
```

```python
from hazm import *
```

```python
from __future__ import unicode_literals
```

```python
file_list = glob.glob(os.path.join(os.getcwd(), "dataset\\datase_khabarvarzeshi\\train", "*.txt"))

train_corpus = []

for file_path in file_list:
    with open(file_path, encoding='utf8') as f_input:
        train_corpus.append(f_input.read())

print(train_corpus)
```

```python
type(train_corpus), len(train_corpus)
```

```python
file_list = glob.glob(os.path.join(os.getcwd(), "dataset\\datase_khabarvarzeshi\\test", "*.txt"))

test_corpus = []

for file_path in file_list:
    with open(file_path, encoding='utf8') as f_input:
        test_corpus.append(f_input.read())

print(test_corpus)
```

```python
type(test_corpus), len(test_corpus)
```

```python
mstr_train = ''
for item in train_corpus:
    mstr_train += str(item) + ' '
#mstr_train
```

```python
mstr_test = ''
for item in test_corpus:
    mstr_test += str(item) + ' '
#mstr_test
```

```python
normalizer = Normalizer()
opt_train = normalizer.normalize(mstr_train)
opt_test = normalizer.normalize(mstr_test)

#opt_test, opt_train
```

```python
opt_train = opt_train.replace('\u200c', '')
opt_train = opt_train.replace('\u200e', '')
opt_train = opt_train.replace('\n', '')
opt_train = re.sub(r'\d+', '', str(opt_train))
#opt_train
```

```python
opt_test = opt_test.replace('\u200c', '')
opt_test = opt_test.replace('\u200e', '')
opt_test = opt_test.replace('\n', '')
opt_test = re.sub(r'\d+', '', str(opt_test))

#opt_test
```

```python
type(opt_test), type(opt_train)
```

```python
stopword_dir = 'C:\\Users\\Farshad\\Desktop\\persian-stopwords-master'
stopwords = []
with open(stopword_dir+'\\persian.txt', encoding='utf8') as words:
    for line in words:
        stopwords.append(line.replace('\n', ''))
```

```python
tokens_train = word_tokenize(opt_train)
len(tokens_train) #tokens_train
```

```python
result_train = [i for i in tokens_train if not i in stopwords]
len(result_train), #result_train
```

```python
tokens_test = word_tokenize(opt_test)
len(tokens_test), #tokens_test
```

```python
result_test = [i for i in tokens_test if not i in stopwords]
len(result_test), #result_test
```

```python
result_train.count('فوتبال'), result_test.count('فوتبال')
```

```python
m_dict = {}
for item in result_train:
    if item not in m_dict:
        m_dict[item] = 1
    else:
        m_dict[item] += 1
len(m_dict), m_dict
```

```python
sort_orders = sorted(m_dict.items(), key=lambda x: x[1], reverse=True)
for i in sort_orders:
    print(i[0], i[1])
```
